```python
import numpy as np
import pandas as pd
import os
import nltk
import gensim
path = os.getcwd()
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf_deleted = pd.read_csv(path + '\\dbow_deleted_clean.csv')
clf_deleted.drop('Unnamed: 0',axis=1,inplace=True)
tfidf = pd.read_csv(path + '\\tfidf.csv')
tfidf.drop('Unnamed: 0',axis=1,inplace=True)
clf_not_deleted = pd.read_csv(path + '\\dbow_not_deleted_clean.csv')
clf_not_deleted.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
X = clf_deleted.iloc[:,1:]
y = clf_deleted['3']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=22)

clf = DecisionTreeClassifier(max_depth=None, min_samples_split=2,random_state=0)
scores = cross_val_score(clf, X_train, y_train, cv=5)
scores.mean()
```
